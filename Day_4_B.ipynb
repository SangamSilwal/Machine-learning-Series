# 1. OneHotEncoding using pandas

```python
df_encoded = pd.get_dummies(df, columns=["fuel", "owner"]) #It will give boolean -> true(1) anb false(0)
df_encoded[df_encoded.select_dtypes(bool).columns] = df_encoded.select_dtypes(bool).astype(int)
```

```python
df_encoded
```

# 2. K-1 OneHotEncoding

```python
df_encoded = pd.get_dummies(df, columns=["fuel", "owner"],drop_first=True) #It will give boolean -> true(1) anb false(0)
df_encoded[df_encoded.select_dtypes(bool).columns] = df_encoded.select_dtypes(bool).astype(int)
```
