```python
from sklearn.datasets import make_classification
import numpy as np
X,y = make_classification(n_samples=100,n_features=2,n_informative=2,n_redundant=0,random_state=41,n_classes=2,n_clusters_per_class=1,hypercube=False,class_sep=10)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

```python
def step(z):
  return 1 if z>0 else 0
```

```python
def perceptron(X,y):
  X = np.insert(X,0,1,axis=1)
  weight = np.ones(X.shape[1])
  lr = 0.1

  for i in range(1000):
    j = np.random.randint(0,100)
    y_hat = step(np.dot(X[j],weight))
    weight = weight + lr*(y[j]-y_hat)*X[j]
  return weight[0],weight[1:]
```

```python
intercept_,coef_ = perceptron(X,y)
```

```python
print(intercept_)
print(coef_)
```
