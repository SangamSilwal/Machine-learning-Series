```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/content/placement.csv")
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,0:-1],df.iloc[:,-1],test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
clf1 = LogisticRegression()
clf2 = DecisionTreeClassifier()
```

```python
clf1.fit(X_train,y_train)
clf2.fit(X_train,y_train)
```

```python
y_pred1 = clf1.predict(X_test)
y_pred2 = clf2.predict(X_test)
```

Now Using Accuracy Score to check Decision tree and logistics Regression

```python
from sklearn.metrics import accuracy_score
print("Logistics Regression Accuracy: ",accuracy_score(y_test,y_pred1))
print("Decision Tree Accuracy: ",accuracy_score(y_test,y_pred2))
```

Using confusion tree in Sklearn

```python
from sklearn.metrics import confusion_matrix
pd.DataFrame(confusion_matrix(y_test,y_pred1))
```

```python
# The confusion Matrix will misguide us when there is imbalanced dataset.
# For Example: -taking binary results i.e 1 or 0
# if 1 is 99999 and 0 is rarely 1-> Then this leads to imbalanced Dataset
```

# **Precision, Recall and F1 Score**

```python
from sklearn.metrics import precision_score,recall_score,f1_score
```

```python
print("For Logistics Regression")
print("Precision_score: ",precision_score(y_test,y_pred1))
print("recall_score: ",recall_score(y_test,y_pred1))
print("f1_score: ",f1_score(y_test,y_pred1))
```

```python
print("For decision tree")
print("Precision_score: ",precision_score(y_test,y_pred2))
print("recall_score: ",recall_score(y_test,y_pred2))
print("f1_score: ",f1_score(y_test,y_pred2))
```

For multidimensional result i.e not binary outputs , the precision, recall and f1 score are bit different. We also have macro precision/Recall and weighted precison and/Recall

```python
print("For Logistics Regression")
print("Precision_score: ",precision_score(y_test,y_pred1,average=None))
print("recall_score: ",recall_score(y_test,y_pred1,average=None))
print("f1_score: ",f1_score(y_test,y_pred1,average=None))
```

```python
print("For Logistics Regression")
print("Precision_score: ",precision_score(y_test,y_pred1,average='macro'))
print("recall_score: ",recall_score(y_test,y_pred1,average='macro'))
print("f1_score: ",f1_score(y_test,y_pred1,average='macro'))
```

```python
# We also have classification Report if we want all this
# Note: This is much sucitable for multidimensional output datasets
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred1))
```
